```python
import os
import shutil

# Define the source root directory
source_root = "/content/drive/MyDrive/Kvasir_Detection_Dataset"

# Iterate over each subfolder in the source root directory
for subfolder in os.listdir(source_root):
    subfolder_path = os.path.join(source_root, subfolder)

    # Check if it's a directory
    if os.path.isdir(subfolder_path):
        # Create a new "merged" folder within the current subfolder
        merged_folder_path = os.path.join(subfolder_path, "merged")
        os.makedirs(merged_folder_path, exist_ok=True)

        # Define paths for images and labels directories within the current subfolder
        images_path = os.path.join(subfolder_path, "images")
        labels_path = os.path.join(subfolder_path, "labels")

        # Copy images from the "images" directory to the "merged" folder
        if os.path.exists(images_path):
            for image_file in os.listdir(images_path):
                src_image_path = os.path.join(images_path, image_file)
                dst_image_path = os.path.join(merged_folder_path, image_file)
                shutil.copy(src_image_path, dst_image_path)

        # Copy labels from the "labels" directory to the "merged" folder
        if os.path.exists(labels_path):
            for label_file in os.listdir(labels_path):
                src_label_path = os.path.join(labels_path, label_file)
                dst_label_path = os.path.join(merged_folder_path, label_file)
                # Check for filename conflicts and resolve them
                if not os.path.exists(dst_label_path):
                    shutil.copy(src_label_path, dst_label_path)
                else:
                    # If a conflict exists, append a suffix to the filename
                    base, extension = os.path.splitext(label_file)
                    new_label_file = f"{base}_label{extension}"
                    dst_label_path = os.path.join(merged_folder_path, new_label_file)
                    shutil.copy(src_label_path, dst_label_path)

print("Copying and merging completed.")
```
